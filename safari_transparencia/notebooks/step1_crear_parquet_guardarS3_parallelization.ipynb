## Descarga de archivos a la S3 y creación de las columnas tipo_archivo_respuesta y cantidad_archivos_respuesta, haciendo paralelización de tareas.

Identificamos con cuantos cores cuenta nuestra máquina para aprovechar el máximo.

```python
cores = multiprocessing.cpu_count()
pool = ThreadPool(cores)
cores
```

```python
df['tipo_archivo_respuesta'], df['cantidad_archivos_respuesta'] = zip(*pool.map(etl.MimeType, df['ARCHIVORESPUESTA']))
pool.close()
pool.join()
```
